# I Single Site Consistency

```python
dim = 2
q = 2.0 * dim
Tc = 2.0 * dim
m_vec = [1.0]
T_vec = [0]
```

```python
for iter in range(-1000, 1000):
    t = iter / float(1100.0)
    beta = (t * Tc + Tc) ** (-1)
    T_vec.append(1.0 / beta)
    m = 0.1
    for iter2 in range(1000):
        m = math.tanh(beta * q * m)
    m_vec.append(m)
```

```python
def approx(T):
    return sqrt(3)*(Tc/T-1)**0.5
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.plot(T_vec, m_vec)

newt=linspace(Tc/1.5,Tc,1000)

ax.plot(newt,list(map(approx,newt)))

ax.set_title('Weiss self-consistency for the Ising model in $d$ = ' +  str(dim) + ' dimensions')
ax.set_xlabel('$T$ (Temperature)', fontsize=18)
ax.set_ylabel('$m$ (magnetization per site)', fontsize=18)

savefig('mean_field_self_consistency_Ising.png')

pylab.show()
```

# II Lattice self-consistency: Abscence of fluctuations

```python
dim = 2 
L = 100
N = L ** dim
Tc = 2 * dim
```

```python
neighbours_all = [[site + ((site // L ** ind + i) % L -
                             (site // L ** ind) % L) * L ** ind
                              for ind in range(dim)
                              for i in [1, -1]]
                              for site in range(N)]
```

```python
t =0.01
beta = (t * Tc + Tc) ** (-1) #t=T/Tc-1
m = [1.0] * N
print ('Temperature = ', 1.0 / beta)
```

```python
for iter in range(1000 * N):
    i = random.randint(0,N)
    if i!= N/2+L/2:
        m[i] =  math.tanh(beta * sum(m[k] for k in neighbours_all[i]))
```

```python
X,Y=meshgrid(arange(L),arange(L))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X,Y,lol(m,L))

savefig('mean_field_wireframe_T='+str(1.0/beta)+'.png')
```

```python
x_vec = list(map(int,range(L)+N/2*ones(L)))
y_vec = [abs(m[k]) for k in x_vec]
```

```python
pylab.title('Mean-field lattice self-consistency, $L, d, T =$ ' + str(L) + ' ' + str(dim) + ' ' + str(1 / beta))
pylab.plot(x_vec, y_vec)
pylab.xlabel('$k$ (Lattice site in $(1,0,0)$ direction) ', fontsize=18)
pylab.ylabel('$m(k)$ (spin magnetization)', fontsize=18)
pylab.legend(loc='lower left')
pylab.savefig('mean_field_gen_d_Ising.png')
pylab.show()
```

# Correlation Length

```python
T=linspace(0.01,0.25,5)
s=[]
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

for t in T:
    beta = (t * Tc + Tc) ** (-1) #t=T/Tc-1
    m = [1.0] * N
    for iter in range(200 * N):
        i = random.randint(1,N)
        m[i] =  math.tanh(beta * sum(m[k] for k in neighbours_all[i]))
    x_vec = range(20)
    y_vec = [abs(m[k]) for k in x_vec]
    ax.semilogy(x_vec,y_vec,label='T='+str(1.0/beta))
    s.append(-1.0/stats.linregress(x_vec,log(y_vec))[0])

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.set_xlabel('lattice site')
ax.set_ylabel('magnetization')
ax.set_title('Correlation length')
plt.show()
savefig('magnetization_decay.png')
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.plot([(t * Tc + Tc) for t in T],s)
ax.set_xlabel('Temperature')
ax.set_ylabel('Correlation length')
ax.set_title('Correlation length')

savefig('correlation_length.png')
```

```python
plot(log(T),log(s))
```

## Critical exponent

```python
nu=stats.linregress(log(T),log(s))[0]
nu
```

# Connected correlation


```python
T=-linspace(0.01,0.25,5)
s=[]
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

for t in T:
    beta = (t * Tc + Tc) ** (-1) #t=T/Tc-1
    m = [1.0] * N
    for iter in range(200 * N):
        i = random.randint(1,N)
        m[i] =  math.tanh(beta * sum(m[k] for k in neighbours_all[i]))
    x_vec = arange(20)
    y_vec = [abs(m[k]) for k in x_vec]-min(absolute(m))*ones(len(x_vec))
    ax.semilogy(x_vec,y_vec,label='T='+str(1.0/beta))
    s.append(-1.0/stats.linregress(x_vec,log(y_vec))[0])

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.set_xlabel('lattice site')
ax.set_ylabel('connected magnetization')
ax.set_title('connected correlation length')
show()
savefig('connected_magnetization_decay.png')
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.plot(T,s)
ax.set_xlabel('Temperature')
ax.set_ylabel('Connected Correlation length')
ax.set_title('Connected Correlation length')

savefig('connected_correlation_length.png')
```

# Coment on critical exponent computation

The computation of the correlation length is computationaly higly expensive. And the temperature shouldn't be pushed closed to $T_c$ as this will generate a huge correlation length, much higher that 20 (The region used for the fitting). Keeping temperatures for which the correlation length is around an order of magnitude smaller than that seems to give a good enough result.
