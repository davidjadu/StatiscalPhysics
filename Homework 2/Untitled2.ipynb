# A) Maximum

```python
def F(l):
    x = sorted(map(int,(- log(random.uniform(0.0, 1.0,1000000)) / l)))
    N=[i for i in x if i >= 14]
    data=[sorted(randint(1,n+1,4))[3] for n in N]
    return [N[i] for i in arange(len(N)) if data[i]==14]
```

```python
Lambda=1.0/10.0
hist=F(Lambda)
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))
pylab.hist(hist, range=(0, max(hist)), bins=max(hist), normed=True)
pylab.title("Bayes approach to the tank problem, $\lambda = $" + str(Lambda)+", efficient" )
pylab.xlabel('$N$')
pylab.ylabel('$p(N)$')
pylab.savefig('bayes_max_efficient.png')
pylab.show()
```

```python
L=1.0/linspace(5,100,20)
data=list(map(F,L))
```

```python
p=[sum(d)/len(d) for d in data]
sqp=[sum(power(d,2))/len(d) for d in data]
s=sqp-power(p,2)
```

```python
fig, ax= plt.subplots(1, 2, figsize=(16, 5))
ax[0].scatter(1/L,p)
ax[0].set_ylabel('$\mu_N$ aposteriori',size=16)
ax[0].set_xlabel('$1/\lambda=\mu_N$ apriori',size=16)
ax[0].set_title('Posterior average',size=16)

ax[1].scatter(1/L,sqrt(s))
ax[1].set_ylabel('$\sigma_N$ aposteriori',size=16)
ax[1].set_xlabel('$1/\lambda=\mu_N$ apriori',size=16)
ax[1].set_title('Posterior standard deviation',size=16)
savefig('posterior_average_variance.png')
plt.show()
```

# B) Total Sample

```python
def FT(l):
    x = sorted(map(int,(- log(random.uniform(0.0, 1.0,10000000)) / l)))
    N=[i for i in x if i >= 14]
    data=[sorted(randint(1,n+1,4)) for n in N]
    return [N[i] for i in arange(len(N)) if all(data[i]==[1,2,3,14])]
```

```python
#Takes about 10 min to run
histT=FT(Lambda)
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))
pylab.hist(histT, range=(0, max(histT)), bins=max(histT), normed=True)
pylab.title("Bayes approach to the tank problem, $\lambda = $" + str(Lambda)+", efficient, Total Sample" )
pylab.xlabel('$N$')
pylab.ylabel('$p(N)$')
pylab.savefig('bayes_max_efficient_Total_Sample.png')
pylab.show()
```

```python
#Takes around 10min to run
L=1.0/linspace(5,100,10)
dataT=list(map(FT,L))
```

```python
pT=[sum(d)/len(d) for d in dataT]
sqpT=[sum(power(d,2))/len(d) for d in dataT]
sT=sqpT-power(pT,2)
```

```python
fig, ax= plt.subplots(1, 2, figsize=(16, 5))
ax[0].scatter(1/L,pT)
ax[0].set_ylabel('$\mu_N$ aposteriori',size=16)
ax[0].set_xlabel('$1/\lambda=\mu_N$ apriori',size=16)
ax[0].set_title('Posterior average',size=16)

ax[1].scatter(1/L,sqrt(sT))
ax[1].set_ylabel('$\sigma_N$ aposteriori',size=16)
ax[1].set_xlabel('$1/\lambda=\mu_N$ apriori',size=16)
ax[1].set_title('Posterior standard deviation',size=16)
savefig('posterior_average_variance_TotalSample.png')
plt.show()
```
