```python
%pylab inline
import math
```

```python
#The natural length of the problem is 2*sigma
sigma=0.5 
```

```python
def Z(N,L):
    return (L-2*N*sigma)**N if L>2*N*sigma else 0
def b(n,k):
    return math.factorial(n)/(math.factorial(k)*math.factorial(n-k))
```

```python
def pi(x,N,L):
    K=arange(0,N)
    return (1.0/Z(N,L))*sum([b(N-1,k)*Z(k,x-sigma)*Z(N-1-k,L-x-sigma) for k in K])
```

```python
N=100
L=101
X=linspace(0,1,1000)
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.plot(X,[pi(L*t,N,L) for t in X])

ax.set_title("$\pi(x)$ for N="+ str(N) + ", L=" + str(L),size=16)
ax.set_xlabel("$x/L$",size=16)
ax.set_ylabel("$\pi(x)$",size=16)
savefig('p(x),N='+str(N)+'L='+str(L)+'.png')
pylab.show()
```

```python
N=15
L=[16,18,20,30,50,100,500,1000]
X=linspace(0,1,1000)
```

```python
fig, ax = plt.subplots(2, 4, figsize=(15,6))
fig.subplots_adjust(wspace=.5,hspace=.5)

for i in arange(len(L)):
    ax[int(i/4)][i%4].plot(X,[L[i]*pi(L[i]*t,N,L[i]) for t in X])
    ax[int(i/4)][i%4].set_ylim(0,10)
    ax[int(i/4)][i%4].plot(X,ones(len(X)))
    ax[int(i/4)][i%4].set_title("$\pi(x)$ for N="+ str(N) + ", L=" + str(L[i]))
    ax[int(i/4)][i%4].set_xlabel("$x/L$")
    ax[int(i/4)][i%4].set_ylabel("$L\pi(x)$")

savefig('p(x),N='+str(N))
pylab.show()
```
