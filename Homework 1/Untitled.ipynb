```python
%pylab inline
import math, scipy.misc
```

```python
n = 10
x_val = []
renyi = []
int_renyi = []
```

```python
for y in range(-200, 201):
    x = y / 10.0
    if abs(x) < n:
        kmax = int ((x + n)/ 2.0)
        renyi_sum = 0.0
        for k in range(0, kmax + 1):
            renyi_sum += (-1) ** k * scipy.misc.comb(n, k) * (n + x - 2 * k) ** (n - 1) /  2 **n / math.factorial(n - 1)
    else:
        renyi_sum = 0.0
    x_val.append(x)
    renyi.append(renyi_sum)
delx = x_val[1] - x_val[0]
```

```python
for k in range(len(renyi)):
    int_renyi.append(sum(renyi[k:]) * delx)
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))
ax.set_ylim(0.0, 1.0)
ax.plot(x_val, renyi, linewidth = 4, label='Renyi')
ax.plot(x_val, int_renyi, linewidth = 4, label='Int Renyi')
ax.plot(x_val[201:],(1.0/3.0)*n*power(x_val[201:],-2),linewidth = 4, label='Chebyshev\'s bound')
ax.plot(x_val[201:],divide((2.0/3.0)*n*ones(len(x_val[201:])),((2.0/3.0)*n*ones(len(x_val[201:]))+power(x_val[201:],2))),linewidth = 4, label='Cantelli\'s bound')
ax.plot(x_val[201:],exp(-1.0/(2.0*n)*power(x_val[201:],2)),linewidth = 4, label="Hoeffding\'s bound")
ax.legend(loc='upper right')
ax.set_title("Renyi Formula for the sum of $n$ = " + str(n) + " indep.random ran(-1,1)")
ax.set_xlabel('x')
ax.set_ylabel('p(x)')
savefig('renyi.png')
pylab.show()
```

```python
Ch=(1.0/3.0)*n*power(x_val[201:],-2)
Ca=divide((2.0/3.0)*n*ones(len(x_val[201:])),((2.0/3.0)*n*ones(len(x_val[201:]))+power(x_val[201:],2)))
H=exp(-1.0/(2.0*n)*power(x_val[201:],2))
m=minimum(Ch,minimum(Ca,H))
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))
ax.set_ylim(0.0, 1.0)
ax.plot(x_val, int_renyi, linewidth = 4, label='Int Renyi')
ax.plot(x_val[201:],m,linewidth=4,label='minimum bound')
ax.legend(loc='upper right')
ax.set_title("Renyi Formula for the sum of $n$ = " + str(n) + " indep.random ran(-1,1)")
ax.set_xlabel('x')
ax.set_ylabel('p(x)')
savefig('renyi_min.png')
plt.show()
```
